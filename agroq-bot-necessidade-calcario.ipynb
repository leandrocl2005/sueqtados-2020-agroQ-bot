```python
import telebot
from telebot import types
import my_infos

token = my_infos.token
bot = telebot.TeleBot(token)


info_user = dict()

info = {
    'CTC': 0,
    'V1': 0,
    'cultura': '',
    'PRNT': ''
}

v2_table = {
    "cafe": 60,
    "cana": 60,
    "soja": 50,
    "milho": 70
}

def calcula_necessidade_calcario(chat_id):
    V2 = v2_table[info[chat_id]["cultura"]]
    V1 = info[chat_id]["V1"]
    CTC = info[chat_id]["CTC"]
    PRNT = info[chat_id]["PRNT"]
    resultado =  (CTC * (V2 - V1)) / PRNT
    return round(resultado, 1)

@bot.message_handler(commands=['start'])
def diga_ola(message):
    chat_id = message.chat.id
    bot.send_message(chat_id, "Olá, tudo bem? Eu sou o AgroQ!")
    bot.send_message(chat_id, "Para calcular a necessidade de calcário digite /calcario")
    
    
@bot.message_handler(commands=['calcario'])
def pedir_CTC(message):
    chat_id = message.chat.id
    info[chat_id] = dict()
    bot.send_message(
        chat_id, 
        "O CTC encontra-se nos resultados da análise do solo."
    )
    bot.send_message(
        chat_id, 
        "Por favor, informe o CTC."
    )
    bot.register_next_step_handler(
        message,
        pedir_v1
    )
    
def pedir_v1(message):
    chat_id = message.chat.id
    info[chat_id] = dict()
    info[chat_id]['CTC'] = float(message.text)
    bot.send_message(
        chat_id, 
        "O v1 também está na análise do solo."
    )
    bot.send_message(
        chat_id, 
        "Por favor, informe o v1."
    )
    bot.register_next_step_handler(
        message,
        pedir_cultura
    )
    
def pedir_cultura(message):
    chat_id = message.chat.id
    info[chat_id]["V1"] = float(message.text)
    
    markup = types.ReplyKeyboardMarkup(row_width=2)
    markup.add(types.KeyboardButton("cafe"))
    markup.add(types.KeyboardButton("milho"))
    markup.add(types.KeyboardButton("soja"))
    markup.add(types.KeyboardButton("cana"))

    msg = bot.send_message(
        chat_id,
        'Qual sua cultura?', 
        reply_markup=markup
    )
    
    bot.register_next_step_handler(msg, pedir_PRNT)
    
def pedir_PRNT(message):
    chat_id = message.chat.id
    info[chat_id]["cultura"] = message.text
    bot.send_message(
        chat_id, 
        "O Poder Relativo de Neutralização Total " + 
        "encontra-se na embalagem do calcário."
    )
    bot.send_message(
        chat_id, 
        "Qual o PRNT?"
    )
    bot.register_next_step_handler(
        message,
        mostrar_calculo_calcario
    )
    
def mostrar_calculo_calcario(message):
    chat_id = message.chat.id
    info[chat_id]["PRNT"] = float(message.text)
    resultado = calcula_necessidade_calcario(chat_id)
    bot.send_message(
        chat_id, 
        "Você precisa de " + str(resultado) + " t/ha."
    )
    bot.send_message(
        chat_id, 
        "Se precisar de uma consultoria personalizada, contate nossa empresa!"
    )
    bot.send_message(
        chat_id, 
        "E-mail: consultoria@agroq.com"
    )
    bot.send_message(
        chat_id, 
        "Fale com Ciclano Beltrano da Fulana."
    )
    
bot.polling()
```
